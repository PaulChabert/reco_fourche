### Code pour générer des fausses données

```python
def fake_matrix(nb_products = 6,
                nb_cust = 5,
                share_empty = 0.7):

    name_products = ["P"+str(x+1) for x in range(nb_products)]
    name_cust = ["cust_"+str(x+1) for x in range(nb_cust)]

    rand = np.random.rand(nb_cust,nb_products)
    empty = np.random.rand(nb_cust,nb_products)>share_empty

    fake_matrix = pd.DataFrame(rand*empty,
                               index=name_cust,
                               columns=name_products
                              )


    return fake_matrix
```

```python
fake_data = fake_matrix(100, 100, 0.7)
display(fake_data)
```

```python
# Imports
from surprise import Reader, Dataset, SVD, NMF
from surprise.model_selection import cross_validate, train_test_split
from surprise.accuracy import rmse, mae
```

Exemple https://towardsdatascience.com/various-implementations-of-collaborative-filtering-100385c6dfe0

Doc scikit-surprise https://surprise.readthedocs.io/en/stable/getting_started.html#load-from-df-example

```python
# On crée un Reader de scikit-surprise
reader = Reader(rating_scale=(1, 5))
```

```python
# On transforme les données des commandes au format que surprise veut
# -> 3 colonnes: colonne 1, le customer, colonne 2, l'item, colonne 3, la note
data = Dataset.load_from_df(orders_scoring[['customer_id', 'sku', 'LO']], reader)
```

```python
data
```

## SVD

```python
# On crée un modèle SVD
model = SVD()
```

```python
# J'essaie de cross-validate, pour voir le score du modèle sans aucune optimisation
%%time
cross_validate(model, data, measures=['RMSE'], cv=5, verbose=True)
```

```python
# On va faire un training à la main.
# On sépare train et test set
trainset, testset = train_test_split(data, test_size=0.25)
```

```python
# On fit le model
model.fit(trainset)
```

```python
# On fait des prédictions
predictions = model.test(testset)
```

```python
# Résultat
rmse(predictions)
```

```python
mae(predictions)
```

```python
# On peut faire des prédictions de score, pour un customer et un produit au hasard
# .est -> renvoie le score attendu
model.predict('h6IvomPZxOTKuKGuvS8b3nZrzVk2', '4-LAV-105')
```
